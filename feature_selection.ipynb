```python
#reset dataframe index to allow rows to be accessed by index
data = data.reset_index(drop = True)
```

```python
#get poster image data in the form of numpy array for analysis
image_data = []
movie_name = []
indicator = []
for m in range(len(data['movie_name'])):
    movie = data['movie_name'][m]
    indic = data['tomato_indic'][m]
    try:
        img_data = cv2.imread("C:/Users/16789/Documents/MovieProject/poster/"+movie.replace("?","").replace("/","")+".jpg")
        if(not img_data is None):
            image_data.append(img_data)
            movie_name.append(movie)
            indicator.append(indic)
    except:
        pass
```

```python
#Extracting the number of faces on a movie poster as a feature, as some drama/documentaries don't have faces on their posters and the movies are rated higher. This is done using the cv2 package
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_frontalface_alt.xml")
```

```python
faces_list = []
for i in range(len(image_data)):
    image = image_data[i]
    gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray_image,minSize = (30,30))
    faces_list.append(len(faces))
```

```python
image_info = pd.DataFrame({"movie_name":movie_name,"num_faces":faces_list})
```

```python
image_info
```

```python
#Merge image info data with main dataset
feature_data = pd.merge(data,image_info,on = 'movie_name',how = 'inner')
```

```python
#analyze movie plots
nlp = spacy.load("en_core_web_sm")

#filter so no n/a and nan in plot
feature_data = feature_data[feature_data['Plot'] != "N/A"]
feature_data = feature_data[feature_data['Plot'].notna()]

#convert all text to single tense, remove punctuation and number
feature_data["plot_no_punct"] = feature_data['Plot'].apply(lambda text: 
                                      " ".join(str(token.lemma_).lower() for token in nlp(text) 
                                               if not token.pos_ in {'PUNCT','NUM'}))
```

```python
feature_data['plot_no_punct']
```

```python
#vectorize plot data
vectorized = CountVectorizer(ngram_range = (1,1),max_df = 0.95).fit(feature_data['plot_no_punct'])
```

```python
#plot vocabulary
vectorized.vocabulary_
```

```python
len(vectorized.vocabulary_)
```

```python
#remove large amounts of names from plots
def remove_names(text):
    filtered_text = []
    doc = nlp(text)
    for x in doc.ents:
        if(not x.label_ == 'PERSON'):
            filtered_text.append(x.text)
            
    return " ".join(filtered_text)

feature_data['plot_no_punct_no_names'] = feature_data['plot_no_punct'].apply(remove_names)
```

```python
#vectorize plot data
vectorized = CountVectorizer(ngram_range = (1,1),max_df=0.95).fit(feature_data['plot_no_punct_no_names'])
```

```python
#plot vocabulary
vectorized.vocabulary_
```

Plot information for the movies won't be used as a feature due to the large number of names and specific information included in each plot. Before removing names the vocabulary of words is very large and would likely cause overfitting. After removing names, many plots were reduced to one or empty words so it would not provide much information.

```python
#remove final n/a and nan
feature_data = feature_data[(feature_data['first_actor'] != "N/A") & (feature_data['first_actor'] != "NaN")]
feature_data = feature_data[(feature_data['first_director'] != "N/A") & (feature_data['first_director'] != "NaN")]
feature_data = feature_data[feature_data['Plot'] != "N/A"]
feature_data = feature_data[feature_data['Runtime'] != "nan"]
```
