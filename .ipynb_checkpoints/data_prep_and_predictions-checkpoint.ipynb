```python
data = pd.read_csv("cleaned.csv")
```

```python
#function to normalize data
def NormalizeData(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))
```

```python
#create column with main actor
first_actor = []
for actor in data['Actor']:
    try:
        first_actor.append(actor.split(",")[0].lstrip())
    except:
        first_actor.append("N/A")
```

```python
#create column with main production
first_prod = []
for prod in data['Production']:
    try:
        first_prod.append(prod.split(",")[0].lstrip())
    except:
        first_prod.append("N/A")
```

```python
image_data = []
movie_name = []
indicator = []
for m in range(len(feature_data['movie_name'])):
    movie = feature_data['movie_name'][m]
    indic = feature_data['tomato_indic'][m]
    try:
        img_data = cv2.imread("C:/Users/16789/Documents/MovieProject/poster/"+movie.replace("?","").replace("/","")+"-grey"+".jpg")
        image_data.append(img_data)
        movie_name.append(movie)
        indicator.append(indic)
    except:
        pass
```

```python
max_height = 0
max_width = 0
for i in image_data:
    max_height = max(max_height,i.shape[0])
    max_width = max(max_width,i.shape[1])
```

```python
WHITE = [255,255,255]
resized_image_data = []
indic_for_train = []
for i in range(len(image_data)):
    height = image_data[i].shape[0]
    width = image_data[i].shape[1]
    constant= cv2.copyMakeBorder(image_data[i],math.floor((max_height-height)/2),math.ceil((max_height-height)/2),math.floor((max_width-width)/2),math.ceil((max_width-width)/2),cv2.BORDER_CONSTANT,value=WHITE)
    if constant.shape == (533,300,3):
        resized_image_data.append(constant)
        if(indicator[i] == 0):
            indic_for_train.append([1,0])
        else:
            indic_for_train.append([0,1])
```

```python
model = Sequential()
model.add(Conv2D(64, kernel_size=(3,3), strides=(1,1), padding='valid', activation='relu', input_shape=(max_height,max_width,3)))
model.add(MaxPool2D(pool_size=(5,5)))
model.add(Conv2D(32,kernel_size = (3,3),strides = (1,1), padding = 'valid',activation = 'relu'))
model.add(MaxPool2D(pool_size = (5,5)))
model.add(Conv2D(26,kernel_size = (3,3),strides = (1,1), padding = 'valid',activation = 'relu'))
model.add(MaxPool2D(pool_size = (2,2)))
model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dense(2,activation = 'softmax'))
```

```python
model.summary()
```

```python
resized_image_data = np.stack(resized_image_data)
indic_for_train = np.stack(indic_for_train)
```

```python
model.compile(loss = 'binary_crossentropy',metrics = ['accuracy'],optimizer = 'adam')
```

```python
model.fit(resized_image_data,indic_for_train)
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_frontalface_alt.xml")
```

```python
faces_list = []
brightness = []
saturation = []
hue = []
brightness_sd = []
saturation = []
saturation_sd = []
hue_sd = []
blue = []
green = []
green_sd = []
red = []
blue_sd = []
red_sd = []
for i in range(len(image_data)):
    image = image_data[i]
    gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray_image,minSize = (30,30))
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    brightness.append(np.mean(hsv[:,:,2]))
    saturation.append(np.mean(hsv[:,:,1]))
    hue.append((hsv[:,:,0]))
    brightness_sd.append(np.std(hsv[:,:,2]))
    saturation_sd.append(np.std(hsv[:,:,1]))
    hue_sd.append(np.std(hsv[:,:,0]))

    blue.append(np.mean(image[:,:,0]))
    green.append(np.mean(image[:,:,1]))
    red.append(np.mean(image[:,:,2]))
    blue_sd.append(np.std(image[:,:,0]))
    green_sd.append(np.std(image[:,:,1]))
    red_sd.append(np.std(image[:,:,2]))
    print(len(faces),movie_name[i],i)
    faces_list.append(len(faces))
```
